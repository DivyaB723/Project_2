```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/ONLINE_FOOD_DELIVERY_ANALYSIS.csv')

print(df.describe())

df.info()
```

```python
# Convert order_date to datetime
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')

# Convert to integer
int_columns = ['Customer_Age']

for col in int_columns:
    df[col] = df[col].fillna(0).astype(int)


# Convert categorical/object columns to 'category'
cat_columns = ['Customer_Gender', 'City', 'Area', 'Payment_Mode']

for col in cat_columns:
    df[col] = df[col].astype('category')

print(df.dtypes)
```

```python
# filling of missing datas

#1 Customer_Gender

df['Customer_Gender'].fillna(df['Customer_Gender'].mode()[0], inplace=True)

#2 Cuisine_Type

df['Cuisine_Type'] = df['Cuisine_Type'].ffill()

#3 Area

df['Area'] = df['Area'].bfill()

#4 City

df['City'] = df['City'].bfill()

#5 Order_Date

df['Order_Date'].fillna(df['Order_Date'].mode()[0], inplace=True)
```

```python
#6 Order_Time

df['Order_Time'].fillna(df['Order_Time'].mode()[0], inplace=True)

#7 Peak_Hour

df['Peak_Hour'].fillna(df['Peak_Hour'].mode()[0], inplace=True)
df['Peak_Hour'] = df['Peak_Hour'].astype(object)

#8 Payment_Mode

df['Payment_Mode'].fillna(df['Payment_Mode'].mode()[0], inplace=True)

#9 Cancellation_Reason

df['Cancellation_Reason'] = df['Cancellation_Reason'].fillna('Unavailable')

df.info()

```

```python
#filling missing values by groupby

#1 Customer_Age colunm

df['Customer_Age'] = df.groupby(['Cuisine_Type', 'Customer_Gender'])['Customer_Age'].transform(lambda x: x.fillna(x.median()))

#2 Delivery_Rating

df['Delivery_Rating'] = df.groupby('Delivery_Partner_ID')['Delivery_Rating'].transform(lambda x: x.fillna(x.median()))

#3 delivery time column
df['Delivery_Time_Min'] = df.groupby('Area')['Delivery_Time_Min'].transform(lambda x: x.fillna(x.median()))

df.info()

```

```python
#filling order value, discount applied, final amount using Iterative Imputer

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

column_to_use =  ['Order_Value', 'Discount_Applied', 'Distance_km', 'Customer_Age']

imputer = IterativeImputer(max_iter = 10, random_state = 42)
df[column_to_use] = imputer.fit_transform(df[column_to_use])

df[['Order_Value', 'Final_Amount', 'Discount_Applied']] = df[['Order_Value', 'Final_Amount', 'Discount_Applied']].clip(lower=0)

df.info()

print(df['Final_Amount'].describe())
print(df['Discount_Applied'].describe())
print(df['Order_Value'].describe())
```

```python
#Correcting invalid values

#1 Profit margin

profit_margin_negative = df[df['Profit_Margin'] < 0]

df['Profit_Margin'] = df['Profit_Margin'].abs()

print(f'negatives_after: {(df['Profit_Margin'] < 0).sum()}')

#2 Delivery_Rating

delivery_outliers = df[df['Delivery_Rating'] > 5]

df.loc[df['Delivery_Rating'] > 5, 'Delivery_Rating'] = 5

#3 Restaurant_Rating

restaurant_outliers = df[df['Restaurant_Rating'] > 5]

df.loc[df['Restaurant_Rating'] > 5, 'Restaurant_Rating'] = 5
```

```python
# Detecting outliers (delivery time, order value, etc)

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"Column: {col} — Found {len(outliers)} outliers")
```

```python
# Capping of outlier in Discount_Applied

df['Discount_Applied'] = df['Discount_Applied'].apply(lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x)

# Correcting Final_Amount column based on the capped outlier

df['Final_Amount'] = df['Final_Amount'].fillna(df['Order_Value'] - df['Discount_Applied'])
```

```python
df.info()
```

```python
# Feature Engineering

#1 Adding Profit_Margin_Percentage column

df['Profit_Margin_Percentage'] = (df['Profit_Margin'] / df['Final_Amount']) * 100

df['Profit_Margin_Percentage'] = df['Profit_Margin_Percentage'].replace([np.inf, -np.inf], 0)

df['Profit_Margin_Percentage'] = df['Profit_Margin_Percentage'].fillna(0)


#2 Peak hour indicator

df['Peak_Hour_Indicator'] = df['Peak_Hour'].map({True: 'Peak', False: 'Non-Peak'})


#3 Customer age groups

conditions = [(df['Customer_Age'] <= 24), (df['Customer_Age'] > 24) &
              (df['Customer_Age'] <= 44), (df['Customer_Age'] > 44) &
              (df['Customer_Age'] <= 60), (df['Customer_Age'] > 60)]

labels = ['Young-Adult', 'Adult', 'Middled-Aged', 'Senior']

df['Customer_Age_Group'] = np.select(conditions, labels, default='Unknown')

print(df['Customer_Age_Group'].value_counts())


#4 Delivery performance categories

conditions = [(df['Delivery_Time_Min'] <= 30), (df['Delivery_Time_Min'] > 30) &
              (df['Delivery_Time_Min'] <= 45), (df['Delivery_Time_Min'] > 45) &
              (df['Delivery_Time_Min'] <= 60), (df['Delivery_Time_Min'] > 60)]

labels = ['Excellent', 'Good', 'Average', 'Poor']

df['Delivery_Performance_Category'] = np.select(conditions, labels, default='Unknown')

print(df['Delivery_Performance_Category'].value_counts())

df['Delivery_Performance_Category'] = df['Delivery_Performance_Category'].astype('category')


#5 Order day type (Weekday / Weekend)

df['Order_Day_Type'] = np.where(df['Order_Date'].dt.dayofweek < 5, 'Weekday', 'Weekend')

print(df['Order_Day_Type'].value_counts())

df.info()
```

```python
df.to_csv("FDA_Cleaned.csv", index= False)
```
